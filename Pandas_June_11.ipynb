```python
columns_to_show = ['Total day minutes', 'Total eve minutes', 'Total night minutes']

# describe() per group
df.groupby(['Churn'])[columns_to_show].describe()
```

```python
pd.crosstab(df['Churn'], df['Customer service calls'])
```

```python
# margins=True adds row/column totals
pd.crosstab(df['Churn'], df['Voice mail plan'], margins=True)
```

```python
df.pivot_table(['Total day calls', 'Total eve calls', 'Total night calls'],
               ['Area code'],
               aggfunc='mean')
```

```python
total_calls = (df['Total day calls'] + df['Total eve calls'] +
               df['Total night calls'] + df['Total intl calls'])

df.insert(loc=len(df.columns), column='Total calls', value=total_calls)
df.head()
```

```python
df['Total charge'] = (df['Total day charge'] + df['Total eve charge'] +
                      df['Total night charge'] + df['Total intl charge'])
df.head()
```

```python
# Remove the two columns we just made. axis=1 means “columns”.
df = df.drop(['Total charge', 'Total calls'], axis=1)
```

```python
# df.drop([1, 2]).head()   # without axis, drop deletes ROWS by index
df.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = (8, 6)
```

```python
pd.crosstab(df['Churn'], df['International plan'], margins=True)

sns.countplot(x='International plan', hue='Churn', data=df)
plt.savefig('int_plan_and_churn.png', dpi=300)
```

```python
pd.crosstab(df['Churn'], df['Customer service calls'], margins=True)

sns.countplot(x='Customer service calls', hue='Churn', data=df)
plt.savefig('serv_calls_and_churn.png', dpi=300)
```

```python
df['Many_service_calls'] = (df['Customer service calls'] > 3).astype('int')

pd.crosstab(df['Many_service_calls'], df['Churn'], margins=True)

sns.countplot(x='Many_service_calls', hue='Churn', data=df)
plt.savefig('many_serv_calls_and_churn.png', dpi=300)
```
